```python
#let1 -- phosphorilated
def compare_confint_probs(smat, let1, let2, alpha):
    n1 = colsum(smat, let1)
    n2 = colsum(smat, let2)
    print("aacid\tp_psite_low\tp_psite\tp_psite_high\tp\tsignificant")
    for k in smat.keys():
        try:
            p1 = smat[let1][k]/n1
            p2 = smat[let2][k]/n2
        except:
            continue
        a = st.binom.interval(alpha, n1, p1, loc=0)
        if p2 > a[0]/n1 and p2 < a[1]/n1:
            sign = '-'
        else:
            if p1 > p2:
                sign = '>'
            else:
                sign = '<'
#            sign = '+'
        outarr = [k, a[0]/n1, p1, a[1]/n1, p2, sign]
        print('\t'.join(map(str, outarr)))
```

```python
def substvec_table_anal(substvec_tab_filename, alpha_one_side):
    psite_muts_dict = dict()
    permut_mut_dict = dict()
    permut_num = 0
    substvec_diff_dict = dict()
    with open(substvec_tab_filename) as substvec_tab:
        for s in substvec_tab:
            s = s.strip().split()
            if s[0] == '#':
                aacid_list = s[1:]
                strlen = len(s)
                for c in aacid_list:
                    permut_mut_dict[c] = []
                continue
            if s[0] == "-1":
                _sum = 0
                for i in range(1, strlen):
                    _sum += eval(s[i])
                for i in range(1, strlen):
                    psite_muts_dict[aacid_list[i - 1]] = eval(s[i])/_sum
                continue
            permut_num += 1
            _sum = 0
            for i in range(1, strlen):
                _sum += eval(s[i])
            for i in range(1, strlen):
                permut_mut_dict[aacid_list[i - 1]].append(float(eval(s[i]))/_sum)
    for c in psite_muts_dict.keys():
        psite_val = psite_muts_dict[c]
        permut_mean = sum(permut_mut_dict[c])/permut_num
        sort_arr = sorted(permut_mut_dict[c])
        permut_low = sort_arr[int(permut_num*alpha_one_side)]
        permut_high = sort_arr[-int(permut_num*alpha_one_side)]
        if psite_val < permut_low:
            sign = '<'
        elif psite_val > permut_high:
            sign = '>'
        else:
            sign = '='
        outarr = map(str, [c, psite_val, permut_low, permut_mean, permut_high, sign])
        substvec_diff_dict[c] = outarr
#        print('\t'.join(outarr))
    return substvec_diff_dict
    
    
```
